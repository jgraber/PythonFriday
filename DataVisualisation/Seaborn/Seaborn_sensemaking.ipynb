## What is the distinguishing feature of these groups?

```python
sns.scatterplot(x='flipper_length_mm', 
                y='bill_depth_mm', 
                hue='sex', 
                data=penguins)
```

```python
sns.scatterplot(x='flipper_length_mm', 
                y='bill_depth_mm', 
                hue='species', 
                data=penguins)
```

```python
sns.pairplot(penguins, hue='species')
#plt.savefig('Seaborn_Penguins_Pair_Colour.png')
```

When you have many overlapping data points in a relplot

```python
sns.swarmplot(x="day", 
              y="tip", 
              hue="sex", 
              data=tips)
```

## Load the Flights data set

```python
flights_monthly = sns.load_dataset("flights")
```

```python
flights_monthly.info()
```

```python
flights_monthly.sample(5)
```

## Heatmaps
To see clusters of data with many numerical values

```python
# Create pivot with years (left -> right) 
# and months (top -> bottom) 
# and passengers as values in the cells
flights = flights_monthly.pivot(index="month", 
                                columns="year", 
                                values="passengers")
```

```python
flights.head(3)
```

```python
# Draw a heatmap
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, 
            annot=True, 
            fmt="d", 
            linewidths=.25, 
            ax=ax)
```
