## Basic imports

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Change the style for a single figure

```python
x = [1, 2, 3, 4]
y = [7, 6, 1, 2]

with plt.style.context('bmh'):
    fig, ax = plt.subplots()
    ax.bar(x,y)
```

```python
%matplotlib inline

fig, ax = plt.subplots()
ax.bar(x,y)
```

## Change the style for all figures in the notebook

```python
plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()
ax.bar(x,y)
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
```
