```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [0, 6, 1, 2]
```

```python
# Create a figure containing 1 row of two axes.
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)  
ax1.plot(x, y)
ax1.set_title("1. axes")
ax2.plot(x, y, color='green')
ax2.set_title("2. axes")
fig.suptitle('Figure 1: A Figure with 2 Axes');
```

```python
# Create a figure containing 1 row of two axes.
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             constrained_layout=True)  
ax1.plot(x, y)
ax1.set_title("1. axes")
ax2.plot(x, y, color='green')
ax2.set_title("2. axes")
ax3.plot(x, y, color='red')
ax3.set_title("3. axes")
ax4.plot(x, y, color='orange')
ax4.set_title("4. axes")

fig.suptitle('Figure 2: A Figure with 4 Axes');
```
