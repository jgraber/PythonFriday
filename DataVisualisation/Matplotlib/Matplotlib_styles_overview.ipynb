## Basic imports

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [5, 5]
```

## Get a list of all styles

```python
for style in plt.style.available:
    print(style)
```

```python
x = [1, 2, 3, 4]
y = [7, 6, 1, 2]

for style in plt.style.available:
    with plt.rc_context({"figure.max_open_warning": len(plt.style.available)}):
        with plt.style.context(style):
            fig, ax1 = plt.subplots()
            ax1.bar(x, y)
            ax1.set_xlabel("days of performance testing")
            ax1.set_ylabel("# bugs found")
            ax1.set_title(f"Style: {style}")
```
