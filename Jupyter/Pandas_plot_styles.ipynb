## Basic imports

```python
import pandas as pd
import numpy as np
```

## Preparations

```python
months = ['January', 'February', 'March', 'April', 
          'May', 'June', 'July', 'August', 'September', 
          'October', 'November', 'December']

temp_high_average = { 
    "Bern" : [3.4, 5.2, 10.3, 14.5, 18.6, 22.5, 24.6, 24.2, 19.4, 14.0, 7.7, 3.8],
    "Oslo" : [-0.4, 0.5, 4.4, 10.1, 16.5, 20.0, 22.3, 20.9, 15.7, 9.4, 3.9, 0.0],
    "Rome":  [11.1, 12.6, 15.2, 18.8, 23.4, 27.6, 30.4, 29.8, 26.3, 21.5, 16.1, 12.6] 
}

df = pd.DataFrame(data=temp_high_average, index=months)
df
```

## Customise your plots

### basic plot from Pandas

```python
df.plot()
```

### Set labels for axes and a title

```python
df.plot(ylabel="Temperature in °C", 
        xlabel="Months", 
        title="Average Day Temperatures")
```

### Hide the legend

```python
df["Bern"].plot(legend=False)
```

### Show the grid

```python
df.plot(grid=True)
```

### Change the line style

```python
df["Bern"].plot(style="g--")
```

### Change the size

```python
df.plot(figsize=(3, 3))
```

### Use Matplotlib styles

```python
import matplotlib.pyplot as plt

with plt.style.context('fivethirtyeight'):
    df.plot(figsize=(8,3))
```

### Use the Matplotlib objects directly

```python
from matplotlib.ticker import MultipleLocator


ax = df.plot(grid=True)
ax.set_ylabel("Temperature in °C")
ax.set_title("Average Day Temperatures")
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.tick_params(axis='x',which='major',rotation=90)
ax.set_xticks(np.arange(len(months)))
ax.set_xticklabels(months)
```

## Box plot

### The usual box plot

```python
df.plot.box()
```

### Changed colours for the box plot

```python
color = {
    "boxes": "DarkGreen",
    "whiskers": "DarkOrange",
    "medians": "DarkBlue",
    "caps": "Gray",
}

df.plot.box(color=color)
```

## Pie chart

```python
data = {'Animals': {'Cats': 65, 'Dogs': 30}}
df = pd.DataFrame(data=data)
df
```

```python
df.plot.pie(y='Animals')
```

### Values in %

```python
df.plot.pie(y='Animals',autopct="%.2f")
```

### Real values

```python
df.plot.pie(y='Animals', 
            autopct=lambda x: '{:.0f}'.format(x*df["Animals"].sum()/100))
```
