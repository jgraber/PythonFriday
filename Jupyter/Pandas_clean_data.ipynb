## Basic imports

```python
import os
data_file = os.path.abspath(os.path.join(os.getcwd(), 'data', 'NetGreenhouseGasEmissions_EEA_sdg_13_10.xlsx'))
export_file = os.path.abspath(os.path.join(os.getcwd(), 'data', 'Greenhouse_cleaned.csv'))
```

```python
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.simplefilter("ignore")
```

## Read Excel

```python
df = pd.read_excel(data_file, sheet_name='Sheet 1', skiprows=8, skipfooter=3, engine='openpyxl');
```

```python
df
```

## Drop columns without values

```python
df = df.dropna(axis=1, how='all')
```

```python
df
```

## Fix headers

```python
headers = df.iloc[0]
```

```python
headers
```

```python
type(headers)
```

```python
headers = headers.astype(str).apply(lambda x: x.replace('.0',''))
headers = headers.apply(lambda x: x.replace('TIME','Country'))
```

```python
df  = pd.DataFrame(df.values[1:], columns=headers)
```

```python
df
```

## Set index to 'Country' column

```python
df = df.set_index('Country')
```

```python
df
```

## Replace name of 'Germany'

```python
df = df.rename(index={'Germany (until 1990 former territory of the FRG)': 'Germany'})
```

```python
df
```

## Remove row with no data

```python
df = df.drop(labels='GEO (Labels)', axis=0)
```

```python
df
```

## Switch axes

```python
df = df.swapaxes(1, 0)
```

```python
df
```

## Save the data frame

```python
df.to_csv(export_file, index=True)
```
