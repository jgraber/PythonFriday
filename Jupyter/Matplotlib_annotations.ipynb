## Basic imports

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Annotate a plot

### Horizontal line

```python
x = [1, 2, 3, 4]
y = [7, 6, 1, 2]
sla = 4

fig, ax = plt.subplots()
ax.bar(x,y)
ax.axhline(sla, linestyle=':', color='orange', linewidth=5)
ax.annotate(f'{sla}', xy=(4.4, 4.1))
```

### Vertical line

```python
data =  [1, 1, 2, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5, 5, 5, 5, 5, 6, 6] 

fig, ax = plt.subplots()
ax.hist(data, bins=6)
ax.axvline(4, linestyle=":", color="green")
ax.annotate('Goal', xy=(4.05, 5.5), color="green")
```

```python
x = [1, 2, 3, 4]
y = [7, 6, 1, 2]
sla = 4

fig, ax = plt.subplots()
ax.bar(x,y)

ax.annotate("Attention!", xy=(3, 1), xytext=(3.5, 4), color="magenta",
            arrowprops=dict(arrowstyle="->", color="red", linewidth=2, 
                            linestyle="-", hatch="x"))
```
