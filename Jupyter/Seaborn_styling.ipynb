### use a palette in your graph

```python
penguins = sns.load_dataset("penguins")
```

```python
sns.pairplot(penguins, hue='species', palette='pastel')
```

### Custom palette 

```python
custom = ["#ffaf3c", "#3c5587", "#adb7c4"]
sns.pairplot(penguins, hue='species', palette=custom)
```
