## Basic imports

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Creating a graph

```python
G = nx.Graph()
```

### Add nodes

```python
G.add_node(1)
```

```python
G.add_nodes_from([2, 3])
```

```python
G.add_node("A")
```

```python
G.add_nodes_from(["B", "C"])
```

```python
G.number_of_nodes()
```

```python
G.nodes()
```

### Add edges

```python
G.add_edge(1, 2)
```

```python
G.add_edges_from([(1, 'A'), (2, 3), ("B", "C")])
```

```python
G.number_of_edges()
```

```python
G.edges()
```

## Draw the graph

```python
options = {
    'node_color': 'deepskyblue',
    'node_size': 400,
    'width': 3,
    'with_labels': True
}
```

```python
nx.draw(G, **options)
```

```python
nx.draw_circular(G, **options)
```

```python
nx.draw_planar(G, **options)
```

```python
nx.draw_shell(G, **options)
```

## Create directed graphs

```python
G = nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_node("A")
G.add_nodes_from(["B", "C"])
G.add_edge(1, 2)
G.add_edges_from([(1, 'A'), (2, 3), ("B", "C")])
```

```python
nx.draw_circular(G, **options)
```
