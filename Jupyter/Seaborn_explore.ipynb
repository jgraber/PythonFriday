## Basic imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import os
data_file = os.path.abspath(os.path.join(os.getcwd(), 'data', 'diamonds.csv'))
```

## Read the data from CSV

```python
df = pd.read_csv(data_file)
```

```python
df = sns.load_dataset("diamonds")
```

```python
df.info()
```

```python
sns.relplot(x="carat", y="price", hue="cut", data=df)
```

```python
sns.stripplot(x="cut", y="price", hue="carat", data=df)
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax)
```

```python
sns.pairplot(df, plot_kws=dict(marker='+', linewidth=1), diag_kws=dict(fill=False))
```
