## Parallel coordinates

```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/pandas-dev/'
    'pandas/main/pandas/tests/io/data/csv/iris.csv'
)
df
```

```python
pd.plotting.parallel_coordinates(
    df, 'Name', color=('#556270', '#4ECDC4', '#C7F464')
)
```
